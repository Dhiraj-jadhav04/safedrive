# Advanced Drowsiness Detection Engine
This notebook tests the core features: EAR, MAR, Head Pose, and rPPG Heart Rate estimation.

```python
import cv2
import time
import matplotlib.pyplot as plt
from utils import DrowsinessDetector

detector = DrowsinessDetector()
cap = cv2.VideoCapture(0)

print("Starting engine test (Press 'q' to stop)...")

while cap.isOpened():
    ret, frame = cap.read()
    if not ret: break
    
    data, results = detector.process_frame(frame)
    
    # Visual overlays
    cv2.putText(frame, f"EAR: {data['ear']:.2f}", (20, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.putText(frame, f"MAR: {data['mar']:.2f}", (20, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.putText(frame, f"BPM: {int(data['bpm'])}", (20, 90), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.putText(frame, f"P: {int(data['pitch'])}, Y: {int(data['yaw'])}, R: {int(data['roll'])}", (20, 120), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    
    cv2.imshow('Drowsiness Engine Test', frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
print("Engine test stopped.")
```
